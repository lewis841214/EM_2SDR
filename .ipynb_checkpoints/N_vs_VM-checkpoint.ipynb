```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
n = 24
n_list = [2,5,7, 10, 15, 22, 24]
n_list.reverse()
for n in n_list:
    it_list = []
    vm_list = []
    exp_name = f'20220110_n_{n}_MPCA_noised_SNR_0.5_half'
    dir_path = f'/home/users/lewis841214/Desktop/1project/CryoEMCode/GPU_CryoEM/tomosipo/TSNE_result/{exp_name}'
    from os import listdir
    from os.path import isfile, join
    onlyfiles = [f for f in listdir(dir_path) if isfile(join(dir_path, f))]

    for file in onlyfiles:

        if len(file.split('-')) == 2:
            #print(file.split('-')[1].split())
            it = float(file.split('-')[1].split()[0])
            vm = float(file.split('-')[1].split()[3].split('j')[0][:-1])
            #print(it, vm)
            it_list.append(it)
            vm_list.append(vm)
        it_list = np.array(it_list)
        vm_list = np.array(vm_list)
        sort = np.argsort(it_list)
        it_list = it_list[sort].tolist()
        vm_list = vm_list[sort].tolist()
    
    plt.plot(it_list, vm_list, label = f'n = {n}')
plt.legend()
plt.title(f'noised 2SDR')
plt.xlabel('iter')
plt.ylabel('v_measure')
plt.show()
```
